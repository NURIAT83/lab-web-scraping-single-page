```python
!pip install bs4 
# we use ! when we are not on terminal
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#The link to the url we are going to use
url="https://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
#Check that we are working with the correct url
url
```

```python
#download html with a get request
response = requests.get(url)
```

```python
response.status_code
```

```python
response.content
```

```python
# parse the element
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
#check how the html looks
soup
```

```python
soup.select("body > div.container > div:nth-child(4) > div.col-xs-12.col-md-8 > h2> li")
```

```python
div = soup.find_all('div', class_= "title-artist")
```

```python
div = soup.find_all('div', class_= "title-artist")
```

```python
print(div)
```

```python
title_artist=soup.select("div.chart-content.col-xs-12.col-sm-8 > p")
```

```python
title_A = []
for li in title_artist:
    try:
 
        title_A.append(li.get_text().split("-")[0])
    except:
        print("NA")
  
print(title_A)
```

```python
#Save title-artist combination as a dataframe
TA=pd.DataFrame(title_A)
print (TA)
```
